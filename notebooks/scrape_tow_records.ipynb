```python
def scrapecity(city):
    city_df = pd.DataFrame()
    print("Scraping towers in {}".format(city))
    
    webcity = Select(driver.find_element_by_name('Cities'))
    webcity.select_by_visible_text(city)
    towers_button = driver.find_element_by_xpath('//*[@id="TOWERS"]/span')
    towers_button.click()
    time.sleep(5)
    
    html_source = driver.page_source
    
    if "Cert #" in html_source:
        towerinfo = BeautifulSoup(html_source, 'lxml')
        tower_table = towerinfo.find("table",{"class":"TransportList"})
        tower_rows = tower_table.findAll(lambda tag: tag.name=='tr')
        tower_rows.pop(0)

        for i in tower_rows:
            thetower = i.findAll('td')
            cmpnumb = thetower[0].text
            certno = thetower[1].text
            cname = thetower[2].text
            dba = thetower[3].text
            caddr = thetower[4].text
            ccity = thetower[5].text
            cstate = thetower[6].text
            ins_telno = thetower[7].text
            authtype = thetower[8].text

            city_df = city_df.append({'Complaints': cmpnumb,
                                    'Cert #': certno,
                                    'Name': cname,
                                    'DBA': dba,
                                    'Address': caddr,
                                    'City': ccity,
                                    'State': cstate,
                                    'Telephone': ins_telno,
                                    'Authority': authtype}, ignore_index=True)
    else:
        pass
    return city_df
```

```python
def scrapecomplaints(city):
    comp_df = pd.DataFrame()
    api_broken = []
    compnumbers = towers_all[towers_all['City'] == city]
    
    for i in compnumbers['Complaints'].unique():
        print(">>> {}".format(i))
        driver = webdriver.Chrome(executable_path='../notebooks/chromedriver.exe')
        driver.get("https://eeaonline.eea.state.ma.us/DPU/Fileroom/Transports")
        time.sleep(5)
        
        webcity = Select(driver.find_element_by_name('Cities'))
        webcity.select_by_visible_text(city)
        time.sleep(2)
        
        towers_button = driver.find_element_by_xpath('//*[@id="TOWERS"]/span')
        towers_button.click()
        time.sleep(5)
    
        complink = driver.find_element_by_link_text(i)
        complink.click()
        time.sleep(3)

        html_source = driver.page_source
        
        if "No Complaints." in html_source:
            driver.close()
            continue
        elif "Api is currently unavailable" in html_source:
            api_broken = api_broken.append(i)
        else:
            compinfo = BeautifulSoup(html_source, 'lxml')
            comp_table = compinfo.find("table",{"class":"detailsTable"})
            comp_rows = comp_table.findAll(lambda tag: tag.name=='tr')
            comp_rows.pop(0)

            for row in comp_rows:
                    thecomp = row.findAll('td')
                    date = thecomp[0].text
                    complaint = thecomp[1].text
                    compcomplete = thecomp[2].text
                    result = thecomp[3].text

                    comp_df = comp_df.append({'ComplaintNo': i,
                                            'Date': date,
                                            'Complaint': complaint,
                                            'Complaint Completed': compcomplete,
                                            'Result': result}, ignore_index=True)
        driver.close()
        time.sleep(1)
    return comp_df
```
